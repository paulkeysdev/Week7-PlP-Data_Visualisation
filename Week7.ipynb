```python
##Line Chart - Trends Over Time
import matplotlib.pyplot as plt
import pandas as pd

# Load data
df = pd.read_csv("owid-covid-data.csv")

# Filter for a specific country (e.g., Kenya)
kenya_df = df[df["location"] == "Kenya"]

# Plot time-series of new cases
plt.figure(figsize=(10, 5))
plt.plot(kenya_df["date"], kenya_df["new_cases"], label="Daily New Cases", color="blue")
plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.title("COVID-19 Daily Cases in Kenya")
plt.legend()
plt.show()
```

```python
#bar
import seaborn as sns

# Group data by continent
continent_avg_cases = df.groupby("continent")["total_cases"].mean().reset_index()

# Plot bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x="continent", y="total_cases", data=continent_avg_cases, palette="coolwarm")
plt.xlabel("Continent")
plt.ylabel("Average Cases")
plt.title("Average COVID-19 Cases per Continent")
plt.show()
```

```python
#Scatter Plot - Relationship Between Two Variables
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["total_vaccinations"], y=df["total_cases"], alpha=0.5, color="green")
plt.xlabel("Total Vaccinations")
plt.ylabel("Total Cases")
plt.title("Vaccinations vs. Cases")
plt.show()
```

```python
#Histogram
plt.figure(figsize=(10, 5))
sns.histplot(df["total_cases"].dropna(), bins=30, color="purple", kde=True)
plt.xlabel("Total Cases")
plt.ylabel("Frequency")
plt.title("Distribution of Total COVID-19 Cases")
plt.show()
```
